#### 2.1. Preparation of the dataset A - Season statistic since 1950:

First of all, we wanted to have for each year since 1950, a single line per player who played. Initially, if a player had changed teams, the dataset contained several lines. We wanted to avoid that and to have for each year and each player one line containing the performances and characteristics of the player during this year.

To do so, given a player and a year, if multiples rows were found we agregate the data with the following methods:
* For attributes that counts game facts like points, fouls, etc. We take the sum on all line, i.e. we sum the points, fouls, etc. that a player has scored in all the team he played for during a year.
* For relative attributes like scoring efficiency or performance metric, we simply took the average. This is not correct obviously because. We can take the example of a player that played 1 game with a team and scoring 2 points with only 1 attempt (his 2pt efficiency is then 1). But this player also played for another team that year having an 0.3 efficiency, taking the mean would give an efficiency of 0.65 which is incredibly high compared to regular statistics. It should be computed again from the number of total shots made over the total of shots attempted. Obviously, it would have been easy to do it on shooting efficiency but some other metrics were trickier to compute, require more processing on data, so we decided to avoid that. Especially as players in that case are quite rare and in general taking the average give reasonable results.

### Predict salary based on player statistics

In order find out what metrics are the most important to assess the quality of a player, we decided to use Machine Learning. One assumption has to made to support this analysis:  player’s salary is a good estimate of player’s performances. This can certainly be discussed as the salary might also depends on player’s fame, team budget, economic context as mentioned before, etc.

Based on this, our idea was to create a Random Forest regressor (1) and assess the variables importance in order to find what metric is the most important. The model was trained using statistics and salaries from 2016 and was used to predict 2017’s salaries. This sounds reasonable as salaries are generally computed based on the previous year’s results and salary. 


#### Data preparation

First we clean the data by removing attributes that are not relevant (we assume that birth date has no real influence on the performance). Then we delete all missing value so the dataset can be fit into the Random Forest.
Two datasets are built, one containing 2016's data and one with 2017's, `salaries2016` and `salaries2017`.

* `salaries2016` will be the training dataset.
* Predictions will be made on `salaries2017`. 

```python
salaries = data_df[data_df['Year'].isin([2016,2017])]# load dataset
```

```python
salaries = salaries.drop(['birth_year', 'birth_month', 'birth_day'], axis=1).dropna() #drop row with missing data

# sklearn can only handle numerical data, therefore categorical columns have to be encoded as integers
categorical_feature_mask = salaries.dtypes==object
categorical_cols = salaries.columns[categorical_feature_mask].tolist()
label_encoder = LabelEncoder()
players2017 = salaries[salaries['Year']==2017]
players2017 = players2017[['Player']].drop_duplicates(['Player'], keep='last')
salaries[categorical_cols] = salaries[categorical_cols].apply(lambda col: label_encoder.fit_transform(col))

# keep only data for the years of interest: 2016 and 2017
salaries2016 = salaries[salaries['Year']==2016].drop(['Year'], axis=1).drop_duplicates(['Player'], keep='last')
salaries2017 = salaries[salaries['Year']==2017].drop(['Year'], axis=1).drop_duplicates(['Player'], keep='last')
salaries2016.head(10)
```

```python
# piece of code used to create two numpy array from dataframe
# X contains the features
# y contains the targets
def separate_labels_features(df):
    y = df['salary'].values
    df_without_label = df.drop(['salary', 'Player'], axis=1)
    X = df_without_label.values
    return X, y
```

```python
# create random forest regressor
forest = RandomForestRegressor(criterion='mse', max_depth=6, n_estimators=100, random_state=randint(1,1000))
```

```python
X_train, y_train = separate_labels_features(salaries2016)
X_valid, y_valid = separate_labels_features(salaries2017)

print("Train dataset size:", X_train.shape, y_train.shape)
print("Test dataset size:", X_valid.shape, y_valid.shape)
```

#### Training

```python
# training on the dataset
forest.fit(X_train, y_train)
```

#### Performance assessment

The model's score is computed on the training and the test datasets to assess the quality of the predictions. Validation score is not very good, however this is not that important as we want to use the predictions to assess the performances not to actually predict the salaries. The predicted salary is meant to be a baseline of the performances. The difference between predicted and actual salary then shows how much a player is overpaid or underpaid given its actual performances. 

```python
salaries_predictions = forest.predict(X_valid)
score_train = forest.score(X_train, y_train)
score_validation = forest.score(X_valid, y_valid)
print('Training score: {:.2f}'.format(score_train))
print('Validation score: {:.2f}'.format(score_validation))
```

#### Players with highest salaries

```python
players2017['Actual salary'] = pd.Series(y_valid, index=players2017.index)
players2017['Predicted salary'] = pd.Series(salaries_predictions, index=players2017.index)
players2017['Relative difference'] =  (players2017['Predicted salary'] - players2017['Actual salary']) / players2017['Actual salary']
```

The model variables can be ranked by their importance: variables that are used the most in the decision process will have a high importance. Based on this criterion, we can show what variables are the more important on the following bar chart.
From this, it can be seen that the average points scored per game is the most important metric by far. Then Win Shares (WS) and Player Efficient Rate (PER) are also important, which was foreseeable as those metrics are designed to assess player’s performances.
Starting year is also import for salary prediction, however, this is not simply correlated to the performances of a player. In general, players' performances follow some kind of parabolla curve with a peak around 30 years old while their salary increase during all their career.

## 4. Genre

We chose the genre and the structure of our visulizations based on [4].

### Genre and structure

To tell the story of *what makes a great NBA player* in a way that is accessible for anyone, we thought that we need to explain to the reader the following points:
* What is NBA ?
* Who are the greatest NBA players (some of them)?
* What differentiates great NBA players from others?
* How to identify the great NBA players, based on their statistics / characteristics?
* How do the best players shoot?

Therefore, as we need to answer several questions that have a natural order, we decided to base our visualization on the **magazine style genre**. We can even consider our work as an article, containing several interactive graphs and ending on a partitionned poster for the 2018 Playoffs shots part.

Inspired by **martini-glass structure** the story we want to tell follow a strict linear path while giving an increasing level of freedom to the reader to let him find more by himself.

### Visual narrative tools used
In this section, we list the visual narrative tools we've used in our article. 

Highlighting:
* Zooming: severals graphs allow the reader to zoom in/out. It helps to isolate some data.

Transition guidance:
* Familiar Objects (but still cuts): for the first 3 graphs, when a parameter changes, it can change the signification of the graph (for example if the attribute displayed changed). So we wanted to mark this change with a cut while using the same colors for the position and players.
* Object Continuity: for the heatmaps, it was important for us to use smooth transition to highlight the differencies between the shot of 2 players.
* Animated transitions: we tried as much as possible to have animated transition, when plots are changed. Plotly.js was maybe not the best way to add transition in our visualizations but it was really more convenient than d3.js to create the plots and their interactions.  

### Narrative structure tools used
In this section, we list the narrative structure tools we've used in our article. 

Ordering:
* Linear: as discussed before, the questions we answer follow a natural order.

Interactivity:
* Hover Highlighting / Details: to give more details to the reader, it was important to implement hover on the different plot. The most interesting one is for the boxplot, the hover gives for each group all the information about the box.
* Selection: as we wanted the reader to be able to learn a lot by himself and our dataset contains a lot of information about the players, we left the user select as much as possible the different parameters of the graphs (attributes, player, year, etc).
* Explicit Instruction: to allow the reader to use the full potential of the different interactions we tried to give explicit instructions on how to use them.
* Default Views: we selected wisely all the default settings of the graphs to highlight the story we want to tell. The reader is able to find more by himself, but the point we found the most interesting is the default one.

Messaging:
* Captions / Headlines: it helps the reader to understand what each visualization is about.
* Introductory Text: it helps the reader to understand why a visualization is important.
* Summary / Synthesis: it helps the reader to understand what the visualization explains.

For the category messaging, we could have added "Accompanying Article" but our whole visualization is an article.

## 5. Visualizations

As explained in the previous section, we decided to tell the story of *What makes a great NBA player* through an article containing interactive visualizations.

The structure of our article is the following:

* Introduction about the NBA
* Dataset overview
* Selection of great NBA players
* Evolution of the average player through time
* Players’ distribution by position and by year
* Player attributes and tendancies
* Salary estimation based on player statistics
    * Salary predictions
    * Variables importance
* 2018 Playoffs: player performances and shots

We begin our article by introducing the NBA, the dataset we used and some of the great NBA players. We let the user select which of these players he wants to highlight for the rest of the article. All players are selected by default.

Then we want to tell that the best players are better than the average one. To do so, we choosed to plot the annual evolution of a selected attribute since 1950 for the average and for the selected players.

However as some attributes can greatly vary depending on the position of the player and compare the best players only with the average player is not enough, we decided to box-plot a selected attribute for each of the 5 positions and for a selected year. From this visualization, we can identify differences between the positions and compare the selected players with the distribution of the selected attribute.

Nevertheless, defining player’s performances cannot be summarized by one metric. So we wanted to let the reader identify links between the different attributes, so we can focus only on some of them, given their importance. Therefore, we decided to plot a scatter-plot of two selected attributes for all the player of a selected year. The selected players are displayed with bigger dot. Moreover, a linear-regression is plotted to highlight the trend between the two attributes. This visualization allow us to affirm that the salary is a good indicator of the performance of a player.

Once it has been stated that salary is a performance indicator, we can try to estimate it using a Random Forest regressor. We decided to display the estimated salaries and the variables importance in order to find what metric is the most important.

Finally, we found interesting to let the user zoom into some players of the 2018 playoffs and allows the reader to compare directly to player, by their statistics and their shots. So we included in our article a partitioned poster about the team involved in the 2018 playoffs and the statistics of their players. This visualization includes a heatmap of the shots of the players that let the user identify from where the best players score.

## 6. Discussion

The difficulty of this assignment was to build a story about NBA players and not only a list of facts. We have done our best to build a real story from everything we have learned through our datasets. We tried to follow a linear path to gradually explain how we could identify the great players, starting from a simple comparison to the average and ending by using machine learning to evaluate the importance of the different attributes. Finally, we are happy to have been able to offer many interactions to the user, despite the magazine style used and a quite author-driven narration (observable format does not give much freedom to do anything else).

We found that Points scored, Win Shares, Player Efficiency Rating, were the attributes that are the most important to predict salaries and therefore performances. However they are not measuring efficiency of the players and they do not take into account the number of games or minutes played. It could be interesting to conduct a similar analysis using modified variables that take this into account. Moreover, we could have look at the performances at different time of a game. Great players are also players that performs the best at the right moment, i.e. at the end of the games. Such study has already been conducted [here](https://github.com/danchyy/Basketball_Analytics/blob/master/Scripts/nba_clutch_stats/Nba%20clutch%20time%20and%20regular%20time%20shooting%20comparison.ipynb).

The section about the 2018 playoffs may seems a little bit out of scope, but we thought it would be very interesting to be able to compare two players side by side with shooting heatmaps. Playoffs were well suited to do such a comparison. These heatmaps show that some players are able to shoot a lot and from anywhere, and those players are often the categorized as 'great players'.

Plotly.js helps us a lot to create the first graphs. We saved a lot of time by using this library. However, with more time we would have preferred to use d3.js in order to have much more personalized graphics, with more detailed hovers, and smoother interactions.
